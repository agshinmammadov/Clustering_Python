# Check the outliers in dataset

```python
sns.boxplot(data = df.iloc[: , 1:]) # we have outliers in dataset
```

```python
columns = list(df.columns)
columns.remove('country')
```

```python
def outlier_handler(data):
    sorted(data)
    Q1, Q3 = np.percentile(data, [25,75])
    IQR = Q3 - Q1
    lower_range = Q1 - (1.5*IQR)
    upper_range = Q1 + (1.5*IQR)
    return lower_range, upper_range
```

```python
for col in columns:
    lower_bound, upper_bound = outlier_handler(df[col])
    df[col] = np.clip(df[col],a_min=lower_bound, a_max=upper_bound)
```

```python
sns.boxplot(data = df)
```

# K-means

```python
#Scaling

scaler = StandardScaler()
df = scaler.fit_transform(df.drop('country', axis = 1))
columns = columns
```

```python
df = pd.DataFrame(df,index = data.country, columns = columns)
df
```

```python
knns = KMeans(n_clusters=3)
knns.fit(df)
KMeans(n_clusters=3)
knns.labels_
```

```python
df_final = data.copy()
df_final['Kmeans labels'] = knns.labels_
df_final
```

```python
df_final['Kmeans labels'].replace({1:'Underdeveloped', 0:'Developing', 2: "Developed"}, inplace = True)
df_final
```

# Hierarichal clustering

```python
Hcluster = AgglomerativeClustering(n_clusters = 3)
Hcluster = Hcluster.fit_predict(df)
Hcluster
```

```python
df_final['Hierarichal clustering labels'] = Hcluster
df_final
```

```python
df_final['Hierarichal clustering labels'].replace({1:'Underdeveloped', 0:'Developing', 2: "Developed"}, inplace = True)
df_final
```
